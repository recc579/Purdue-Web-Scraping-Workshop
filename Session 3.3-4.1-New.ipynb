The website we will be using for this part is https://www.getcompanyinfo.com/industry/information-technology-services/

```python
import pandas as pd
from bs4 import BeautifulSoup as bs
import urllib.request
import time
import random
import re
```

```python
url = "https://www.getcompanyinfo.com/industry/information-technology-services/"
html = urllib.request.urlopen(url).read().decode('utf-8')
soup = bs(html)
```

```python
soup
```

First we want to be able to get the company name and the link to the company information.

```python
soup.find_all(title = "Company Name")
```

```python
for a in soup.find_all(title = "Company Name"):
    print(a.get_text())
    print(a["href"])
    print("----")
```

When we click onto a specific company, it turns out that the url base that we want to use is only "https://www.getcompanyinfo.com", not "https://www.getcompanyinfo.com/industry/information-technology-services/"

```python
Companies = [a.get_text() for a in soup.find_all(title = "Company Name")]
Links = [a["href"] for a in soup.find_all(title = "Company Name")]
print(Companies)
print(Links)
```

```python
url_base = "https://www.getcompanyinfo.com"
n=2
html2 = urllib.request.urlopen(url_base+Links[n]).read().decode('utf-8')
soup2 = bs(html2)
```

```python
print(soup2.prettify())
```

```python
# Get the company type
soup2.find("label", text="Company Type:").find_next_sibling().get_text()
```

```python
# Get the industry
soup2.find("label", text="Industry:").find_next_sibling().get_text()
```

```python
# Get when it was founded
soup2.find("label", text="Founded:").find_next_sibling().get_text()
```

```python
# Get the country
soup2.find("label", text="Country:").find_next_sibling().get_text()
```

```python
# Get the headquarters city and state/providence if applicable
hold = soup2.find("label", text="Headquarter:").find_next_sibling().get_text()
split = hold.split(",")
city = split[0]
state = split[1].strip()
print("City:",city)
print("State:",state)
```

```python
# Get the phone number
soup2.find("label", text="Phone:").find_next_sibling().get_text()
```

```python
# Get the email
soup2.find("label", text="Email:").find_next_sibling().get_text()
```

```python
# Get the website
soup2.find("label", text="Website:").find_next_sibling().get_text()
```

```python
# Get the number of employees
soup2.find("h4", text="Employees").find_next_sibling().get_text()
```
